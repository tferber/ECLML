```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
def showercore(r, alpha, beta):
    w = np.exp((r**alpha)/beta)
#     return 1./w
    return w
```

```python
s = np.linspace(0,0.4,100)
```

```python
ws = showercore(r=s, alpha=3, beta=0.075)
```

```python
plt.scatter(s, ws)
```
