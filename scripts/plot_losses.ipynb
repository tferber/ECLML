```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('niceplots')
```

```python
dict_train = np.load('/nfs/dust/belle2/user/ferber/github-ECLML/saved_model/20210313-165816-04164/dict_train-final.npy', allow_pickle=True)[()]
dict_test = np.load('/nfs/dust/belle2/user/ferber/github-ECLML/saved_model/20210313-165816-04164/dict_test-final.npy', allow_pickle=True)[()]
```

```python
epochs = []
train_losses = []
test_losses = []
train_times = []
test_times = []
lrs = []

train_mean_1 = []
train_mean_2 = []
train_mean_bkg = []
train_sigma_1 = []
train_sigma_2 = []
train_sigma_bkg = []
train_a_1 = []
train_a_2 = []
train_a_bkg = []

test_mean_1 = []
test_mean_2 = []
test_mean_bkg = []
test_sigma_1 = []
test_sigma_2 = []
test_sigma_bkg = []
test_a_1 = []
test_a_2 = []
test_a_bkg = []

nepochs = len(dict_train)
for idx in range(1, nepochs):
    epochs.append(idx)
    train_losses.append(dict_train[idx][0])
    test_losses.append(dict_test[idx][0])
    train_times.append(dict_train[idx][1])
    test_times.append(dict_test[idx][1])
    lrs.append(dict_train[idx][3])
    
    train_mean_1.append(dict_train[idx][4][0])
    train_sigma_1.append(dict_train[idx][5][0])
    train_a_1.append(dict_train[idx][6][0])
        
    train_mean_2.append(dict_train[idx][4][1])
    train_sigma_2.append(dict_train[idx][5][1])
    train_a_2.append(dict_train[idx][6][1])
    
    train_mean_bkg.append(dict_train[idx][4][2])
    train_sigma_bkg.append(dict_train[idx][5][2])
    train_a_bkg.append(dict_train[idx][6][2])
    
    
    test_mean_1.append(dict_test[idx][4][0])
    test_sigma_1.append(dict_test[idx][5][0])
    test_a_1.append(dict_test[idx][6][0])
        
    test_mean_2.append(dict_test[idx][4][1])
    test_sigma_2.append(dict_test[idx][5][1])
    test_a_2.append(dict_test[idx][6][1])
    
    test_mean_bkg.append(dict_test[idx][4][2])
    test_sigma_bkg.append(dict_test[idx][5][2])
    test_a_bkg.append(dict_test[idx][6][2])    
```

```python
plt.plot(epochs, train_losses, 'o', ms=2,  color='red', label='train')
plt.plot(epochs, test_losses, 'o',  ms=2,  color='black', label='test')
plt.gca().set_yscale('log')
plt.gca().set_xlabel(r'Epoch')
plt.gca().set_ylabel(r'Loss / graph')
plt.xlim(0,nepochs)
plt.ylim(5.5e-4,6e-4)
plt.yscale('log')
plt.legend()
```

```python
plt.plot(epochs, lrs, 'o', ms=2, color='black')
plt.gca().set_xlabel(r'Epoch')
plt.gca().set_ylabel(r'Learning rate')
plt.xlim(0,nepochs)
```

```python
plt.plot(epochs, train_mean_1, 'o', ms=2, color='green', label='training, k=1')
plt.plot(epochs, train_mean_2, 'o', ms=2, color='blue', label='training, k=2')
plt.plot(epochs, train_mean_bkg, 'o', ms=2, color='orange', label='training, k=bkg')
plt.gca().set_xlabel(r'Epoch')
plt.gca().set_ylabel(r'Mean')
plt.xlim(0,nepochs)
plt.legend()
```

```python
ms = 20
plt.scatter(epochs, train_a_1,c='green', s=ms, label='training, k=1')
plt.scatter(epochs, train_a_2, c='blue', s=ms, label='training, k=2')
plt.scatter(epochs, train_a_bkg, c='orange', s=ms, label='training, k=bkg')
plt.scatter(epochs, test_a_1, edgecolors='green', s=ms, facecolors='none', label='test, k=1')
plt.scatter(epochs, test_a_2, edgecolors='blue', s=ms, facecolors='none', label='test, k=2')
plt.scatter(epochs, test_a_bkg, edgecolors='orange', s=ms, facecolors='none', label='test, k=bkg')
# plt.scatter(epochs, train_a_2, 'o', ms=2, edgecolors='blue', label='training, k=2')
# plt.scatter(epochs, train_a_bkg, 'o', ms=2, edgecolors='orange', label='training, k=bkg')
# plt.scatter(epochs, test_a_1, 'o', ms=2, facecolors='none', edgecolors='green', label='test, k=1')
# plt.scatter(epochs, test_a_2, 'o', ms=2, facecolors='none', edgecolors='blue', label='test, k=2')
# plt.scatter(epochs, test_a_bkg, 'o', ms=2, facecolors='none', edgecolors='orange', label='test, k=bkg')
plt.gca().set_xlabel(r'Epoch')
plt.gca().set_ylabel(r'a')
plt.xlim(0,nepochs)
plt.ylim(0.94, 0.97)
plt.legend()
```

```python
plt.plot(epochs, train_sigma_1, 'o', ms=2, color='green', label='training, k=1')
plt.plot(epochs, train_sigma_2, 'o', ms=2, color='blue', label='training, k=2')
plt.plot(epochs, train_sigma_bkg, 'o', ms=2, color='orange', label='training, k=bkg')
plt.gca().set_xlabel(r'Epoch')
plt.gca().set_ylabel(r'$\sigma$')
plt.xlim(0,nepochs)
plt.legend()
```
